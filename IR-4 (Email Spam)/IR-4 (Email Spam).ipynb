```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
data = pd.read_csv('emailsq.csv', encoding='latin-1')
data
```

```python
data = data.drop(columns=['Email No.'])
```

```python
data['message'] = data.drop(columns=['Prediction']).apply(lambda x: ' '.join(x.astype(str)), axis=1)
```

```python
 data = data[['Prediction', 'message']]
```

```python
 data.columns = ['label', 'message']
```

```python
 data['label'] = data['label'].map({0: 0, 1: 1})
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['message'], data['label'], test_size=0.2, random_state=42)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
 model = MultinomialNB()
```

```python
model.fit(X_train_tfidf, y_train)
```

```python
 y_pred = model.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
